```python
import pandas as pd
import operator
import numpy as py
```

```python
df = pd.read_csv("../globalterrorismdb.csv", encoding = "latin-1", low_memory=False, 
                 usecols=['eventid', 'iyear','iday','success','nkill', 'nkillus', 'nwound', 'nwoundus','country',
                          'country_txt','region','region_txt','city','latitude','longitude',
                          'attacktype1','attacktype1_txt','targtype1','targtype1_txt','natlty1',
                          'natlty1_txt','weaptype1','weaptype1_txt', 'weapdetail'])
df.dropna(subset=['nkill'], inplace = True)
df.drop(df[df['nkill'] == 0].index , inplace=True)
df['location'] = df[['latitude', 'longitude']].values.tolist()
df.head(20)
```

```python
reg = df[['region', 'region_txt']]
reg_num = dict(zip(reg.region, reg.region_txt))
reg_num
```
